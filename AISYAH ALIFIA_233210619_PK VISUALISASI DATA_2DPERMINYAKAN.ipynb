# Line chart kelulusan tiap usia

#### line chart ini berisi jumlah kelulusan tiap usia, jumlah yang lulus tiap usia adalah usia 20 = 6 orang, 21 = 4 orang, 22 = 7 orang, 23 = 3 orang, dan 24 = 2 orang. Setelah coding line chart dibuat, didapatkan line chart seperti dibawah ini.

```python
import matplotlib.pyplot as plt

# contoh datanya
usia = [20, 21, 22, 23, 24]
jumlah_lulus = [6, 4, 7, 3, 2]

# grafiknya
plt.figure(figsize=(8,4))
plt.plot(usia, jumlah_lulus, marker='o', color='b', linestyle='-')
plt.title('Line chart kelulusan tiap usia')
plt.xlabel('Usia')
plt.ylabel('Jumlah lulus')
plt.grid(True)
plt.show()
```

# Bar chart usia berdasarkan gender

#### Bar chart ini berisi usia berdasarkan gender, usia berdasarkan gender adalah 20 = 4 orang, 21 = 4 orang, 22 = 3 orang, 23 = 3 orang, 24 = 3 orang. Setelah coding Bar chart dibuat, didapatkan Bar chart seperti dibawah ini.

```python
import matplotlib.pyplot as plt

# contoh datanya
usia = ['20', '21', '22', '23', '24']
gender = [4, 4, 3, 3, 3 ]

# grafiknya
plt.figure(figsize=(8,4))
plt.bar(usia, gender, color='b')
plt.title('Bar chart usia berdasarkan gender')
plt.xlabel('usia')
plt.ylabel('gender')
plt.grid(False)
plt.show()
```

# Pie chart gender berdasarkan usia

#### Bar chart ini berisi usia berdasarkan gender, usia berdasarkan gender adalah 20 = 4 orang, 21 = 4 orang, 22 = 3 orang, 23 = 3 orang, 24 = 3 orang. Setelah coding Bar chart dibuat, didapatkan Bar chart seperti dibawah ini.

```python
import matplotlib.pyplot as plt

# Data contoh
usia = ['20', '21', '22', '23', '24']
gender = [4, 4, 3, 3, 3 ]

#grafiknya
plt.figure(figsize=(4, 4))
plt.pie(gender ,labels=usia, autopct="%1.1f%%", startangle=140)
plt.title('Pie Chart gender per usia')
plt.axis('equal') # Agar Lingkaran berbentuk Lingkaran sempurna
plt.show()
```
